```python
import pandas
import datetime
```

```python
colNames = [ "dateStr", "timeStr", "MLAT", "MLT",\
            "vel", "azims", "stIds", "poesMLAT1", "poesMLAT2" ]
esDF = pandas.read_csv( "../data/east-saps-north-2011-2014.txt",\
                       delim_whitespace=True, header=None,\
                      names=colNames)
esDF.head()
```

```python
# get counts of datetime measurements
esCntDate = esDF[ esDF["MLT"] > 12. ].groupby( ["dateStr", "timeStr" ] ).size().reset_index()
esCntDate.columns = [ "dateStr", "timeStr", "cnt" ]
esCntDate = esCntDate.sort( ["cnt"], ascending=False ).reset_index(drop=True)
esCntDate.head(20)
```
